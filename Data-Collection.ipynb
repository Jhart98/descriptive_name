### PushShift

```python
import requests
import pandas as pd
```

```python
url = 'https://api.pushshift.io/reddit/search/submission'
params = {
    'subreddit': 'soccer',
    'limit': 1000,
    'filter': 'subreddit, title, created_utc'
}
```

```python
res = requests.get(url,params)
res.status_code
```

```python
res.json()['data']
```

```python
soccer = pd.DataFrame(res.json()['data'])
soccer.head()
#soccer.to_csv('data/soccer.csv', encoding = 'utf-8', index=False)
```

```python
params2 = {
    'subreddit': 'worldcup',
    'limit': 1000,
    'filter': 'subreddit, title, created_utc'
}
```

```python
res2 = requests.get(url,params2)
res2.status_code
```

```python
world_cup = pd.DataFrame(res2.json()['data'])
world_cup.head()
#world_cup.to_csv('data/world_cup.csv', encoding = 'utf-8', index=False)
```
